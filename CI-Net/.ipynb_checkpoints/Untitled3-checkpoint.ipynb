```python
device = 'cuda:0'

text_raw = ["a person holding a large fish by a serene lake", 
            "a happy black and white dog with perked ears"]

image_raw = [Image.open("0048.png"), Image.open("11370.png")]

processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

#print(processor.image_processor)

processor.image_processor.do_convert_rgb = True
processor.image_processor.do_rescale = True

inputs = processor(text=text_raw, 
                    images=image_raw, 
                    return_tensors="pt", 
                    padding=True,
                    truncation=True).to(device)

x = inputs['pixel_values']
#transform = T.Resize((224, 224))
#x = transform(x)
```

```python
from peft import get_peft_model, LoraConfig

model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
lora_config = LoraConfig(
                        r=64,
                        lora_alpha=32,
                        target_modules=["q_proj", "v_proj"],  # 应用于所有q_proj和v_proj层
                        lora_dropout=0.,
                        bias="none",
                        init_lora_weights=False,
                                )
model = get_peft_model(model, lora_config).to(device)
```

```python
inputs_gen = inputs

inputs_gen['pixel_values'] = x_gen
```

```python
model.zero_grad()

outputs = model(**inputs)
with torch.backends.cuda.sdp_kernel(enable_flash=False, 
                                    enable_math=True, enable_mem_efficient=False):                
    logits_per_image = outputs.logits_per_image
    logits_per_text = outputs.logits_per_text
                            
    batch_size = logits_per_image.shape[0]
    labels = torch.arange(batch_size, device=device)
                            
    loss_img = torch.nn.functional.cross_entropy(logits_per_image, labels)
    loss_txt = torch.nn.functional.cross_entropy(logits_per_text, labels)
    loss = (loss_img + loss_txt) / 2

    grad_parameters = [p for p in model.parameters() if p.requires_grad]
    origin_dy_dx = torch.autograd.grad(loss, grad_parameters)
```

```python
print(len(origin_dy_dx))

print(origin_dy_dx[0].shape)
```

```python
model.zero_grad()

outputs = model(**inputs_gen)
with torch.backends.cuda.sdp_kernel(enable_flash=False, 
                                    enable_math=True, enable_mem_efficient=False):                
    logits_per_image = outputs.logits_per_image
    logits_per_text = outputs.logits_per_text
                            
    batch_size = logits_per_image.shape[0]
    labels = torch.arange(batch_size, device=device)
                            
    loss_img = torch.nn.functional.cross_entropy(logits_per_image, labels)
    loss_txt = torch.nn.functional.cross_entropy(logits_per_text, labels)
    loss = (loss_img + loss_txt) / 2

    grad_parameters = [p for p in model.parameters() if p.requires_grad]
    fake_dy_dx = torch.autograd.grad(loss, grad_parameters)
```

```python
print(origin_dy_dx[0])

print(fake_dy_dx[0])
```

```python
import torch.nn.functional as F

for i in range(96):
    sim = F.cosine_similarity(origin_dy_dx[i].view(-1), fake_dy_dx[i].view(-1), dim=0)
    
    print(sim)
```
