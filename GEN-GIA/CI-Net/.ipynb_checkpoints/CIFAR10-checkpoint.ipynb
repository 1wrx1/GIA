```python
%load_ext autoreload
%autoreload 2

from __future__ import print_function

import argparse
import time
import torch
import torch.nn as nn
import torch.nn.parallel

import torch.utils.data
import matplotlib.pyplot as plt
import warnings 

from vit_pytorch import SimpleViT

from torchvision import datasets, transforms
from model.model_architectures import construct_model
from utils.reconstructed import convert_relu_to_sigmoid
from utils.config import config
from model.generator import trainer
warnings.filterwarnings("ignore")
```

```python
act = 'relu'

exp_i = 0
```

```python
cifar100_mean = [0.5071598291397095, 0.4866936206817627, 0.44120192527770996]
cifar100_std = [0.2673342823982239, 0.2564384639263153, 0.2761504650115967]

dataset = datasets.CIFAR100(
        root='/root/autodl-tmp/cifar100', download=True,
        transform=transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize(cifar100_mean, cifar100_std),
        ]))
```

```python
size_batch = 64

idx = 48

label_all, used_idx, imgs = [], [], []

while len(label_all) < 64:
    img, label = dataset[idx]
    #print(label)

    if label not in label_all or len(label_all) >= 100:
        label_all.append(label)
        used_idx.append(idx)
        imgs.append(img)
        
    idx += 1
        
#print(used_idx)
x = torch.stack(imgs)
```

```python
import matplotlib.pyplot as plt

# 假设你的张量是'tensor'
# 随机生成一个张量 (16, 3, 32, 32)，作为示例
tensor = x.clone().detach()

dm = torch.as_tensor(cifar100_mean)[:, None, None]
ds = torch.as_tensor(cifar100_std)[:, None, None]

tensor = tensor.mul_(torch.as_tensor(ds)).add_(torch.as_tensor(dm))

# 将张量从 (16, 3, 32, 32) 转换为 (16, 32, 32, 3) 以适应 Matplotlib
tensor = tensor.permute(0, 2, 3, 1)

# 创建一个图形窗口和子图 (4x4 网格，因为有16个图像)
fig, axes = plt.subplots(8, 8, figsize=(8, 8))

# 迭代地显示每个图像
for i, ax in enumerate(axes.flat):
    # 将每个图像的像素值缩放到 0-1 范围，因为有可能张量的值超出这个范围
    img = (tensor[i] - tensor[i].min()) / (tensor[i].max() - tensor[i].min())
    
    # 显示图像
    ax.imshow(img)
    ax.axis('off')  # 隐藏坐标轴

# 显示图像网格
plt.show()
```

```python
import torchvision.models as models


# Initialize the manual seed and model
#arch = 'ResNet18'
#model, _ = construct_model(arch, seed=42, num_classes=10, num_channels=3)
'''model = model = SimpleViT(
            image_size=32,
            patch_size=16,
            num_classes=10,
            dim=192,
            depth=12,
            heads=3,
            mlp_dim=768
        )'''
model = models.resnet18(pretrained=False)

model.conv1 = nn.Conv2d(3, 64, 3, 1, 1)
model.maxpool = nn.Identity()
model.fc = nn.Linear(in_features=512, out_features=100, bias=True)

model = model.to(config['device'])

if act == 'relu':
    pass
elif act == 'sigmoid':
    convert_relu_to_sigmoid(model)
```

```python
print(model)
```

```python
#size_batch = 64
config["nz"] = 128
config['total_img'] = size_batch
config['b_size'] = size_batch
config['num_epochs'] = 2000
config['rep_freq'] = config['num_epochs']/100

attack =  {'method':'CI-Net','lr':0.001}
config['lr']=attack['lr']
trainer = trainer(config,attack,model,dataset, used_idx[size_batch*exp_i:size_batch*(exp_i+1)])        

reconstructed_image,avg_score = trainer.attack_training()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(avg_score["0"]['CI-Net']['ssim']['score'])
plt.ylabel("SSIM")
plt.xlabel("Round")
plt.show()
```

```python
import metrics
import lpips


lpips_scorer = lpips.LPIPS(net="alex")

x_rec = reconstructed_image["0"]["image"]['CI-Net']['ssim']["timeline"][0][-1]

std = torch.tensor(cifar100_std).view(3, 1, 1)

test_psnr = metrics.psnr(x_rec, x[size_batch*exp_i:size_batch*(exp_i+1)], factor=1 / std)
test_ssim = metrics.cw_ssim(x_rec, x[size_batch*exp_i:size_batch*(exp_i+1)], scales=5)
lpips_score = lpips_scorer(x_rec, x[size_batch*exp_i:size_batch*(exp_i+1)], normalize=True)

print(test_psnr)
print(test_ssim)
print(torch.mean(lpips_score))
```

```python
print(x_rec.shape)

dm = torch.as_tensor(cifar100_mean)[:, None, None]
ds = torch.as_tensor(cifar100_std)[:, None, None]

print(ds.shape)

if x_rec.min() < -0.3:
    x_rec = x_rec.mul_(torch.as_tensor(ds)).add_(torch.as_tensor(dm))
```

## Results Visualization

```python
# Plot obtained images
import numpy as np
horizontal = int(np.sqrt(size_batch))
vertical = int(np.sqrt(size_batch))
fig, axes = plt.subplots(horizontal, vertical, figsize=(6,6))
for i in range(horizontal):
    for j in range(vertical):
        axes[i,j].imshow(torch.clamp(reconstructed_image["0"]["image"]['CI-Net']['ssim']["timeline"][0][-1][horizontal*i+j],min = 0 ,max =1)\
                         .detach().cpu().permute(1,2,0))
        axes[i,j].axis('off')
plt.show()
```

```python
import os

tensor = x_rec.clone().detach().cpu()


# 创建保存图像的目录
output_dir = f'output_images/cifar_bs{size_batch}_' + act +f'/exp_{exp_i}'
os.makedirs(output_dir, exist_ok=True)

# 定义一个转换器，将tensor转为PIL Image
to_pil_image = transforms.ToPILImage()

# 迭代地保存每个图像
for i in range(tensor.size(0)):
    img_tensor = tensor[i]
    
    # 将每个图像的像素值缩放到 0-1 范围
    #img_tensor = (img_tensor - img_tensor.min()) / (img_tensor.max() - img_tensor.min())
    
    # 转换为 PIL 图像
    img = to_pil_image(img_tensor)
    
    # 保存图像为PNG格式
    img.save(os.path.join(output_dir, f'image_{i+1}.png'))

print(f"Images saved to directory: {output_dir}")
```
