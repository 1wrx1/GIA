```python
gt_64 = gt[48:48+64].cpu()
```

```python
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

s = torch.tensor(std).view(3, 1, 1)

dm = torch.as_tensor(mean)[:, None, None]
ds = torch.as_tensor(std)[:, None, None]
```

```python
rec_64_norm = (rec_64 - dm) / ds
gt_64_norm = (gt_64 - dm) / ds
```

```python
import lpips
import metrics

lpips_scorer = lpips.LPIPS(net="alex")

ssim_score = metrics.cw_ssim(rec_64_norm, gt_64_norm, scales=5)
psnr_score = metrics.psnr(rec_64_norm, gt_64_norm, factor=1 / s)
lpips_score = lpips_scorer(rec_64_norm, gt_64_norm, normalize=True)

print(f'ssim:{ssim_score}')
print(f'psnr:{psnr_score}')
print(f'lpips:{lpips_score.mean()}')
```

```python
def grid_plot(tensor):
    tensor = tensor.clone().detach()
    #tensor.mul_(ds).add_(dm).clamp_(0, 1)
    num_images = 64
    if num_images == 1:
        fig, axes = plt.subplots(1, 1, figsize=(1, 1))
    elif num_images in [4, 8]:
        fig, axes = plt.subplots(1, num_images, figsize=(num_images, num_images))
    else:
        fig, axes = plt.subplots(num_images // 8, 8, figsize=(12, num_images // 16 * 3))
    axes = np.reshape(axes, -1)
    for im, ax in zip(tensor, axes):
        ax.imshow(im.permute(1, 2, 0).cpu())
        ax.axis('off')
    plt.subplots_adjust(wspace=0.02, hspace=0.02)
    plt.subplots_adjust(left=0, right=1, bottom=0, top=1)
    os.makedirs('saved_img') if not os.path.exists('saved_img') else None
    
    file_name = f'saved_img/rec_lenet_bs1_imagenet64.pdf'
    plt.savefig(file_name)

grid_plot(rec_64)
```

```python
dataset = datasets.CIFAR10(
        root='/grp01/saas_medml/runxi/cifar10', download=True, train=False,
        transform=transforms.Compose([
            transforms.ToTensor(),
            #transforms.Normalize(cifar10_mean, cifar10_std),
        ]))

idx = 48

label_all, used_idx, imgs = [], [], []

#psnrs, ssims, lpipss = [], [], []

while len(label_all) < 64:
    img, label = dataset[idx]
    #print(label)

    label_all.append(label)
    used_idx.append(idx)
    imgs.append(img)
        
    idx += 1
        
#print(used_idx)
x = torch.stack(imgs)

print(x.shape)
```

```python
labels = torch.as_tensor(label_all).to('cuda')

print(labels)
```

```python
from models.resnet import resnet20

#torch.manual_seed(0)

x = x.to('cuda')

net = resnet20(10).to("cuda")
compress_rate = 0.5
torch.manual_seed(1234)
net.apply(weights_init)
'''net = models.resnet18(pretrained=False)
net.conv1 = nn.Conv2d(3, 64, 3, 1, 1)
net.maxpool = nn.Identity()
net.fc = nn.Linear(in_features=512, out_features=10, bias=True)'''

net = net.to('cuda')

#net.apply(weights_init)

#compress_rate = 0.01

net = net.eval()

#model_size = 269722

loss_fn = nn.CrossEntropyLoss()

grads = []

torch.manual_seed(0)

for i in range(64):
    #net.zero_grad()
    
    image = x[i]
    if len(image.shape) == 3:
        image = image.unsqueeze(0)

    label = torch.as_tensor(label_all[i], device='cuda')
    if len(label.shape) == 0:    
        label = label.unsqueeze(0)
    #print(label.shape)

    net.zero_grad()
    
    outputs = net(image)
    #print(outputs.shape)
    
    loss = loss_fn(outputs, label)

    dy_dx = torch.autograd.grad(loss, [para for para in net.parameters() if para.requires_grad])
    xs = torch.cat(list((_.detach().clone().view(-1) for _ in dy_dx)))
    #loss.backward()
    
    #xs = torch.cat([para.grad.view(-1) for para in net.parameters()])

    xs = xs.unsqueeze(0)

    #print(xs.shape)

    hash_dim = int(model_size * compress_rate)
    shape = (model_size, hash_dim)
    hash_bin = torch.randint(0, hash_dim, (model_size,))
    idx = torch.cat([torch.arange(model_size).long().unsqueeze(0), hash_bin.unsqueeze(0)], dim=0)
    hashed_matrix = torch.sparse_coo_tensor(idx, torch.ones(model_size), shape).to('cuda')

    #print(hashed_matrix.shape)
    #print(xs.shape)
    
    selected_para = None

    xs = torch.sparse.mm(hashed_matrix.t(), xs.t()).t().contiguous()

    grads.append(xs)

grads_tensor = torch.cat(grads)

print(grads_tensor.shape)
    #print(xs.shape)
#rec = grad_to_img_net(x.view(64, -1))

#print(rec.shape)
```
